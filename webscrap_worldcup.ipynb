```python
r=requests.get("https://www.britannica.com/sports/Cricket-World-Cup")
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.text,"html.parser")
```

```python
results=soup.find_all("td")
```

```python
records=[]
for res in results:
    x=res.text.strip()
    records.append(x)
```

```python
print(records)
```

```python
table=[]
for i in records:
    o=records[0:6]
    table.append(o)
    del records[0:6]
```

```python
print(table)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(table,columns=["Year","Winner","Score","Runner-up","Score of RU","Result"])
```

```python
df.head()
```

```python
df.to_csv("Worldcup.csv",index=False)
```
